```python
import re

# Function to perform custom replacement in a file and generate a new file as output
def replace_pattern_in_file(input_file, output_file, pattern):
    try:
        with open(input_file, 'r') as input_f:
            text = input_f.read()

            # Perform custom replacement based on the pattern
            new_text = re.sub(pattern, custom_replacement, text)

        with open(output_file, 'w') as output_f:
            output_f.write(new_text)

        print("Replacement completed. New file generated:", output_file)
    except FileNotFoundError:
        print("Input file not found.")
    except PermissionError:
        print("Permission denied.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Custom replacement logic
def custom_replacement(match):
    matched_text = match.group(0)

    if matched_text == "#~:SIR_IbnIshaq_BEG":
        return "HSNXV##P###Letter_BEG_WZATB"
    elif matched_text == "#~:SIR_IbnIshaq_END":
        return "HSNXV##P###Letter_END_WZATB"
    else:
        return matched_text

# File paths and pattern input
input_file = "\test.txt"
output_file = "\new.txt"
pattern = r'#~:SIR_IbnIshaq_(BEG|END)'

# Replace the pattern in the input file using custom replacement logic and generate the output file
replace_pattern_in_file(input_file, output_file, pattern)
```
